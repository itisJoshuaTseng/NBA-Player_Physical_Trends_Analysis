```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取數據
df = pd.read_csv("nba_team_summary.csv")

# 建立圖表
fig, ax1 = plt.subplots(figsize=(12,6))

# 左側 Y 軸 - 平均身高
ax1.set_xlabel("Team Name")
ax1.set_ylabel("Avg Height (cm)", color="blue")
ax1.plot(df["Team Name"], df["Avg Height (cm)"], marker="o", color="blue", label="Avg Height (cm)")
ax1.tick_params(axis="y", labelcolor="blue")

# 右側 Y 軸 - 平均體重
ax2 = ax1.twinx()
ax2.set_ylabel("Avg Weight (lbs)", color="red")
ax2.plot(df["Team Name"], df["Avg Weight (lbs)"], marker="s", color="red", label="Avg Weight (lbs)")
ax2.tick_params(axis="y", labelcolor="red")

# 標題與美化
plt.title("Average Height & Weight of Top 12 Western Conference NBA Teams")
ax1.set_xticklabels(df["Team Name"], rotation=45, ha="right")
fig.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 讀取數據
df = pd.read_csv("nba_team_summary.csv")

# 添加 Win Percentage 欄位
df["Win Percentage"] = [0.83, 0.63, 0.63, 0.63, 0.62, 0.58, 0.58, 0.55, 0.50, 0.48, 0.46, 0.43]

df.to_csv("nba_team_summary.csv", index=False)

# 提取球員的體重數據
weight_columns = [col for col in df.columns if "Weight_lbs" in col]
weight_data = df[weight_columns].T
weight_data.columns = df["Team Name"]

# 繪製體重熱圖
plt.figure(figsize=(12,6))
sns.heatmap(weight_data, cmap="coolwarm", annot=False)
plt.title("Weight Distribution Heatmap")
plt.xticks(rotation=45)
plt.show()

# 進行體重與勝率的關聯分析
# 繪製體重 vs. 勝率散點圖
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Avg Weight (lbs)"], y=df["Win Percentage"], hue=df["Team Name"], s=100, palette="coolwarm")
plt.xlabel("Average Weight (lbs)")
plt.ylabel("Win Percentage")
plt.title("Correlation Between Team Weight and Win Percentage")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()

# 繪製勝率對上體重的熱圖
plt.figure(figsize=(12,6))
sns.heatmap(df.set_index("Team Name")[weight_columns].T.corrwith(df["Win Percentage"]).to_frame(), cmap="coolwarm", annot=True, cbar=True)
plt.title("Win Percentage vs. Weight Heatmap")
plt.xlabel("Win Percentage")
plt.ylabel("Weight Correlation")
plt.show()
```
